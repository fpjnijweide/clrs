```python
import pandas as pd

df = pd.read_csv('clrs_paper_results.csv')
with open("results.txt") as myfile:
    results_txt = myfile.read()
```

```python
df
```

```python
networks = ["GAT","MPNN","PGN","GAT"]
networks_code_list = ["gatv2", "mpnn", "pgn","gat"]
networks_colors = ["red","blue","green","maroon"]

for network in networks[:-1]:
    df[f"{network} Upper Bound"] = df[network] + df[f"{network} STD"]
    df[f"{network} Lower Bound"] = df[network] - df[f"{network} STD"]

networks[0]="GATv2"
df
```

```python
import re
import numpy as np

algorithm_names = list(df['Algorithm'])

algorithms_code_list = [
    'articulation_points',
    'activity_selector',
    'bellman_ford',
    'bfs',
    'binary_search',
    'bridges',
    'bubble_sort',
    'dag_shortest_paths',
    'dfs',
    'dijkstra',
    'find_maximum_subarray_kadane',
    'floyd_warshall',
    'graham_scan',
    'heapsort',
    'insertion_sort',
    'jarvis_march',
    'kmp_matcher',
    'lcs_length',
    'matrix_chain_order',
    'minimum',
    'mst_kruskal',
    'mst_prim',
    'naive_string_matcher',
    'optimal_bst',
    'quickselect',
    'quicksort',
    'segments_intersect',
    'strongly_connected_components',
    'task_scheduling',
    'topological_sort',
]

memory_types = ["NTM","DeQue"]

memory_sizes= [[20,100]  ,[20,100] ]

results_lines=results_txt.splitlines()



```

```python
df
```

```python
GAT_BEST = [
    'dfs',
    'jarvis_march',
    'kmp_matcher',
    'lcs_length',
    'quickselect',
    'task_scheduling'
]
# MPNN algos
MPNN_BEST = [
    'articulation_points',
    'activity_selector',
    'bfs',
    'bridges',
    'dijkstra',
    'graham_scan',
    'mst_kruskal',
    'mst_prim',
    'naive_string_matcher',
    'segments_intersect',
    'strongly_connected_components',
]
PGN_best = [
    'activity_selector',
    'bellman_ford',
    'binary_search',
    'dag_shortest_paths',
    'find_maximum_subarray_kadane',
    'floyd_warshall',
    'matrix_chain_order',
    'minimum',
    'mst_prim',
    'optimal_bst',
    'quickselect',
    'strongly_connected_components',
    'task_scheduling',
    'topological_sort',
]

import matplotlib.pyplot as plt

marker_list=['o','x','^','d']


for algorithm_index in range(30):
    plt.figure(algorithm_index,figsize=(8, 6), dpi=80) # Here's the part I need
    algorithm_name = algorithm_names[algorithm_index]
    plt.title(algorithm_name)
    # plt.ylim([0,100])
    algorithm_code = algorithms_code_list[algorithm_index]
    if algorithm_code in GAT_BEST:
        yy = [df['GAT Upper Bound'][algorithm_index], df['GAT Lower Bound'][algorithm_index]]
        plt.hlines(yy,xmin=0,xmax=100,colors=['red'],linestyles=['dotted'],label='GAT (target)')

    if algorithm_code in MPNN_BEST:
        yy = [df['MPNN Upper Bound'][algorithm_index], df['MPNN Lower Bound'][algorithm_index]]
        plt.hlines(yy,xmin=0,xmax=100,colors=['blue'],linestyles=['dotted'],label='MPNN (target)')

    if algorithm_code in PGN_best:
        yy = [df['PGN Upper Bound'][algorithm_index], df['PGN Lower Bound'][algorithm_index]]
        plt.hlines(yy,xmin=0,xmax=100,colors=['green'],linestyles=['dotted'],label='PGN (target)')

    for network_idx in range(len(networks_code_list)):
        network_name = networks[network_idx]
        network_code = networks_code_list[network_idx]
        network_color = networks_colors[network_idx]
        for memory_idx in range(len(memory_types)):
            memory_type=memory_types[memory_idx]
            memory_size_list = memory_sizes[memory_idx]
            marker = marker_list[memory_idx]
            x = []
            x_last = []
            y = []
            y_last = []
            for memory_size in memory_size_list:
                string_type_1 = f"{algorithm_code}_best_{network_code}_{memory_type}_{memory_size}"
                string_type_2 = f"{algorithm_code}_{network_code}_{memory_type}_{memory_size}_best"
                string_type_last = f"{algorithm_code}_{network_code}_{memory_type}_{memory_size}_last"
                resulting_lines = [line for line in results_lines if (string_type_1 in line or string_type_2 in line or string_type_last in line)]
                last_lines = [line for line in results_lines if (string_type_last in line)]
                if len(resulting_lines)>0:
                    best_line = resulting_lines[0]
                    p = re.compile(r"'score': (.*?),")
                    best_score = p.findall(best_line)[0]
                    col_name = f"{network_name} {memory_type} {memory_size} Best"
                    # if col_name not in df:
                    #     df[col_name] = np.nan
                    # df.loc[algorithm_index, col_name] = best_score
                    x.append(memory_size)
                    y.append(float(best_score)*100)
                if len(last_lines)>0:
                    last_line=last_lines[0]
                    p = re.compile(r"'score': (.*?),")
                    last_score = p.findall(last_line)[0]
                    col_name = f"{network_name} {memory_type} {memory_size} Last"
                    # if col_name not in df:
                    #     df[col_name] = np.nan
                    # df.loc[algorithm_index,col_name] = last_score
                    x_last.append(memory_size)
                    y_last.append(float(last_score)*100)
            # print(x,y)
            if len(x)>0:
                plt.scatter(x,y,marker=marker,s=50,color=network_color,label=f"{network_name} {memory_type}")
                plt.scatter(x_last,y_last,marker=marker,s=25,color=network_color)
    plt.xlabel("Memory size")
    plt.ylabel("Performance in %")
    plt.legend()
    plt.show()

# TODO legend
# TODO figure size
```
